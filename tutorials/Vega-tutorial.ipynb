## Using VEGA to analyze REACTOME pathway activity in PBMC
In this tutorial, we will demonstrate a use case for VEGA on single-cell PBMCs from Kang et al. The dataset is composed of 2 conditions (cells stimulated with interferon-beta and control cells). We will analyze the GMVs (Gene Module Variables) activities of the two populations.

```python
# Load VEGA and packages
import vega
import torch
import scanpy as sc
import numpy as np
import pandas as pd
```

### Loading the data
We will load the PBMC demonstration data. The data were already preprocessed. We also load the REACTOME pathway file as a dictionary, and turn it into a mask to initialize the VEGA model.

```python
# Read Kang PBMC data - densify expression matrix
adata = vega.data.pbmc()
adata.X = adata.X.A
# Load GMT file and make mask for VEGA - we recommend to always add 1 additional unannotated GMV.
reactome_dict = vega.utils.read_gmt('../vega/data/reactomes.gmt', min_g=0, max_g=1000)
n_unannotated = 1
mask = vega.utils.create_pathway_mask(feature_list=adata.var.index.tolist(),
                                      dict_pathway=reactome_dict,
                                      add_missing=n_unannotated,
                                      fully_connected=True)
```

### Initializing VEGA model
We initialize VEGA with the mask we just created from the GMT file. We also constrain decoder's weights to positive weights to aid interpretability when running differential activity procedure later.

```python
model = vega.VEGA(mask,
                  dropout=0.5,
                  beta=0.00005,
                  positive_decoder=True)
```

### Preparing data and training the model
Now that we initialized the model, we can train it. We wrap the data in a Dataloader and train the model for a maximum of 300 epochs. The training stops when the loss function does not improve for 10 epochs (train_patience).

```python
train_loader = vega.utils.prepare_anndata(adata, batch_size=64, shuffle=True)
epoch_history = model.train_model(train_loader,
                                  learning_rate=1e-4,
                                  n_epochs=300,
                                  train_patience=10,
                                  test_patience=None,
                                  save_model=False)
```

### UMAP embedding and GMV activity visualization
VEGA embedding can be used as input to the UMAP visualization algorithm. We project the cells in the latent space and compute the UMAP coordinates with the Scanpy API.

```python
# Get GMV activities/embedding from trained model
adata.obsm['X_vega'] = model.to_latent(torch.Tensor(adata.X)).detach().numpy()
# Compute 2D UMAP embedding of cells using VEGA latent space
sc.pp.neighbors(adata, use_rep="X_vega", knn=15)
sc.tl.umap(adata, min_dist=0.5)
```

```python
sc.pl.umap(adata, color="condition", frameon=False)
```

We can now visualize the activity of the different pathways directly on the UMAP plot. We extract the activity of the interferon $\alpha/\beta$ pathway as a positive control, since this pathway is relevant to the stimulation studied in this dataset.

```python
# Visualizing activity of interferon alpha beta pathway
pathway_list = list(reactome_dict.keys()) + ['UNANNOTATED_'+str(k) for k in range(n_unannotated)]
gmv_df = pd.DataFrame(data=adata.obsm['X_vega'], index=np.arange(adata.shape[0]), columns=pathway_list)
adata.obs['interferon_alpha_beta'] = gmv_df['REACTOME_INTERFERON_ALPHA_BETA_SIGNALING'].values
```

```python
# Visualize interferon activity on UMAP plot
sc.pl.umap(adata, color='interferon_alpha_beta', cmap='seismic', frameon=False)
```

### Differential activity between groups
In the following, we will proceed to perform a differential activity procedure between the control and stimulated groups to see if the difference in interferon activity is significant. We will call a pathway significantly differentially activited if the absolute value of the $\log_e$(Bayes Factor) (K) is greater than 3. 


_Note: The volcano plot will require the adjustText package installed to work properly._

```python
# Separate Anndata by condition
adata_ctrl = adata.copy()[adata.obs['condition']=='control',:]
adata_stim = adata.copy()[adata.obs['condition']=='stimulated',:]
# Run Bayesian differential procedure
res = model.bayesian_diff_exp(adata_stim, adata_ctrl,
                              n_samples=5000, use_permutations=True,
                              n_permutations=10000, random_seed=0)
```

```python
vega.plotting.volcano_plot(res, pathway_list, sig_lvl=3.0, metric_lvl=3.0,
         to_plot={'REACTOME_INTERFERON_ALPHA_BETA_SIGNALING':'INTERFERON_ALPHA_BETA_PATHWAY'},
         title='Stimulated vs. Control')
```
